```python
def convert_ht(height_str):
    # Split the string into feet and inches
    feet, inches = map(int, height_str.split("'"))

    # Convert feet to inches and add to inches
    total_inches = feet * 12 + inches

    # Convert total inches to float
    height_float = float(total_inches)

    return height_float
data['height'] = data["height"].apply(lambda x:convert_ht(x))
data['height'] = data['height'].astype('float64')
```

# Data Preprocessing

```python
data.isna().sum()
```

```python
data.columns
```

```python
#filling the missing values in categorical columns
data['gender'].fillna(data['gender'].mode()[0],inplace=True)
data['strong_foot'].fillna(data['strong_foot'].mode()[0],inplace=True)
data['coaching'].fillna(data['coaching'].mode()[0],inplace=True)
```

```python
data.isna().sum()
```

```python
cols = ['ball_controlling_skills','penalties_conversion_rate','shot_accuracy','matches_played', 'fitness_rating','years_of_experience']
```

```python
#filling the missing values of skewed cols by median
for i in cols:
    data[i].fillna(data[i].median(),inplace=True)
    
```

```python
colu = ['weight','jumping_skills','mental_strength','behaviour_rating']
```

```python
#filling the missing values of normally distributed by mean
for i in colu:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
num_cols = ['weight','ball_controlling_skills', 'body_reflexes', 'body_balance','jumping_skills', 'penalties_conversion_rate', 'mental_strength',
       'goalkeeping_skills', 'defending_skills', 'passing_skills',
       'dribbling_skills', 'shot_accuracy', 'body_strength_stamina',
       'max_running_speed','behaviour_rating',
       'matches_played', 'fitness_rating', 'trophies_won', 'dedication_level','years_of_experience', 'no_of_disqualifications',
       'selection']
for i in num_cols:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
```

```python
Q1 = np.percentile(data['matches_played'],25,method='midpoint')
Q2 = np.percentile(data['matches_played'],50,method ='midpoint')
Q3 = np.percentile(data['matches_played'],75,method = 'midpoint')
```

```python
IQR = Q3 - Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('min is ',low_lim)
print('max is ',up_lim)
```

```python
outlier = []
for x in data['matches_played']:
    if ((x>up_lim) or (x<low_lim)):
        outlier.append(x)
        
len(outlier)        
```

```python
ind1 = data['matches_played']<low_lim
ind1s =data.loc[ind1].index
ind1s

ind2 = data['matches_played']>up_lim
ind2s = data.loc[ind2].index
ind2s
```

```python
data.drop(ind2s,inplace=True)
data.shape
```

# Train test split

```python
X = data.drop(['player_id','selection','name','country'],axis=1)
y = data['selection']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
modellr = lr.fit(X_train,y_train)
y_predlr = modellr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_test,y_predlr)
```

```python
print('Accuracy =', accuracy_score(y_test,y_predlr))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []
neighbors = np.arange(3,15)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
    classifier.fit(X_train,y_train)
    y_predknn = classifier.predict(X_test)
    acc = accuracy_score(y_test,y_predknn)
    metric_k.append(acc)
```

```python
metric_k
```

```python
plt.plot(neighbors, metric_k, 'o-')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors=14, metric = 'euclidean')
classifier.fit(X_train,y_train)
y_predknn = classifier.predict(X_test)
acc = accuracy_score(y_test,y_predknn)
metric_k.append(acc)
```

```python
print(classification_report(y_test,y_predknn))
```

```python
print(accuracy_score(y_test,y_predknn))
```
