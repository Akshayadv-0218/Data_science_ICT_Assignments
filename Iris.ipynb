```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.hist(df['SL'],color = 'red',rwidth = 0.7)
plt.title('SL vs Frequency')
plt.xlabel('SL')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(df['SW'],color = 'red',rwidth = 0.7)
plt.title('SW vs Frequency')
plt.xlabel('SW')
plt.ylabel('Frequency')


```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.hist(df['PL'],color = 'blue',rwidth = 0.7)
plt.title('PL vs Frequency')
plt.xlabel('PL')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(df['PW'],color = 'blue',rwidth = 0.7)
plt.title('PW vs Frequency')
plt.xlabel('PW')
plt.ylabel('Frequency')
```
